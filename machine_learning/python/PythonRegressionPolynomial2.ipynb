```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

npoints = 20

np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, npoints)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, npoints)

x=np.sort(x)
y=np.sort(y)

df = pd.DataFrame() 
df['x'] = x
df['y'] = y

#print(x)
#print(y)

p = np.polyfit(df.x,df.y, 3) # El último parámetro es el grado polinomial
#print(p)

f = np.poly1d(p) # f(x)
print(f)

fig = plt.figure()
df.plot.scatter(x='x', y='y')
plt.plot(x,f(x),color="red")
plt.show()

cc = np.corrcoef(df.x, df.y)[1,0]
mse = mean_squared_error(y, f(x))
#rmse = np.sqrt(mse) # Penaliza con una magnitud mayor a los errores grandes

print('CC: ', cc)
print('MSE: %f' % mse)
#print('RMSE %f' % rmse)

y_pred = f(-4)
print("predicción: ", y_pred)

#for i in x:
#  print(i, f(i))
  
```
